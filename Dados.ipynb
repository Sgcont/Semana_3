```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("mt_cars_new.csv")

#Calculo das mdias de todos os carros inclusos
media_mpg = sum(df["mpg"])/len(df["mpg"])
print("Media MPG:", media_mpg)

media_hp = sum(df["hp"])/len(df["hp"])
print("Media HP:", media_hp)

media_wt = sum(df["wt"])/len(df["wt"])
print("Media WT:", media_wt)

media_qsec = sum(df["qsec"])/len(df["qsec"])
print("Media QSEC:", media_qsec)

media_vs = sum(df["vs"])/len(df["vs"])
print("Media VS:", media_vs)

media_am = sum(df["am"])/len(df["am"])
print("Media AM:", media_am)

media_gear = sum(df["gear"])/len(df["gear"])
print("Media GEAR:", media_gear)

media_carb = sum(df["carb"])/len(df["carb"])
print("Media CARB:", media_carb)

print("")

#Aqui estou defininindo o nome dos carros com maiores e menores mpg, hp, wt, qsec, vs, am e gear
nome_carro_maior_mpg = df.loc[df['mpg'].idxmax(), 'Unnamed: 0']
print(f"Carro com maior mpg: {nome_carro_maior_mpg}")

nome_carro_menor_mpg = df.loc[df['mpg'].idxmin(), 'Unnamed: 0']
print(f"Carro com menor mpg: {nome_carro_menor_mpg}")

nome_carro_maior_hp = df.loc[df['hp'].idxmax(), 'Unnamed: 0']
print(f"Carro com maior hp: {nome_carro_maior_hp}")

nome_carro_menor_hp = df.loc[df['hp'].idxmin(), 'Unnamed: 0']
print(f"Carro com menor hp: {nome_carro_menor_hp}")

nome_carro_maior_wt = df.loc[df['wt'].idxmax(), 'Unnamed: 0']
print(f"Carro com maior wt: {nome_carro_maior_wt}")

nome_carro_menor_wt = df.loc[df['wt'].idxmin(), 'Unnamed: 0']
print(f"Carro com menor wt: {nome_carro_menor_wt}")

nome_carro_maior_qsec = df.loc[df['qsec'].idxmax(), 'Unnamed: 0']
print(f"Carro com maior qsec: {nome_carro_maior_qsec}")

nome_carro_menor_qsec = df.loc[df['qsec'].idxmin(), 'Unnamed: 0']
print(f"Carro com menor qsec: {nome_carro_menor_qsec}")

nome_carro_maior_vs = df.loc[df['vs'].idxmax(), 'Unnamed: 0']
print(f"Carro com maior vs: {nome_carro_maior_vs}")

nome_carro_menor_vs = df.loc[df['vs'].idxmin(), 'Unnamed: 0']
print(f"Carro com menor vs: {nome_carro_menor_vs}")

nome_carro_maior_am = df.loc[df['am'].idxmax(), 'Unnamed: 0']
print(f"Carro com maior am: {nome_carro_maior_am}")

nome_carro_menor_am = df.loc[df['am'].idxmin(), 'Unnamed: 0']
print(f"Carro com menor am: {nome_carro_menor_am}")

nome_carro_maior_gear = df.loc[df['gear'].idxmax(), 'Unnamed: 0']
print(f"Carro com maior gear: {nome_carro_maior_gear}")

nome_carro_menor_gear = df.loc[df['gear'].idxmin(), 'Unnamed: 0']
print(f"Carro com menor gear: {nome_carro_menor_gear}")

print("")

x = df['mpg']  # Varivel para o eixo x (pode ser outra varivel)
y = df['hp']   # Varivel para o eixo y (pode ser outra varivel)
sizes = df['wt'] * 100  # Tamanho dos crculos (multiplicado por 100 para melhor visualizao)

# Criando o grfico
plt.figure(figsize=(30, 18)) # Define o tamanho da figura
plt.scatter(x, y, s=sizes, alpha=1) # Cria o scatter plot
plt.xlabel('mpg') # Rtulo do eixo x
plt.ylabel('hp') # Rtulo do eixo y
plt.title('Grfico de Crculo do mpg pela hp onde o tamanho do circulo diz o wt') # Ttulo do grfico

# Adicionando rtulos para os carros
for i, txt in enumerate(df['Unnamed: 0']):
    plt.annotate(txt, (x[i], y[i]))

plt.show() # Exibe o grfico

print("O grfico acima demonstra o mpg de cada carro no eixo x e o hp de cada carro no eixo y, a espessura do circulo determina o wt ser maior")


```
